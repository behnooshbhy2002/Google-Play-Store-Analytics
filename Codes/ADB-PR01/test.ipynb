```python
df.drop_duplicates(subset=['App Name'], keep='first', inplace=True)
```

```python
null_count = df.isnull().sum()
for column, count in null_count.items():
    print(f"Column '{column}' has {count} null fields.")
```

```python
df['Installs'] = df['Installs'].str.replace(',', '').str.replace('+', '').astype(float)
```

```python
df.head(20)
```

```python
df['Price'] = df['Price'].astype(float)
```

```python
df.head(20)
```

```python
df['Rating Count'] = df['Rating Count'].fillna(0).astype(int)
```

```python
def convert_size(size):
    if pd.isna(size) or size == 'Varies with device':
        return None
    size = size.replace(',', '')  
    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'k' in size:
        return float(size.replace('k', '')) / 1024
    return None
```

```python
df['Size'] = df['Size'].apply(convert_size)
```

```python
df['Released'] = pd.to_datetime(df['Released'], errors='coerce')
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df['Category'] = df['Category'].str.lower().str.strip()
```

```python
df['Content Rating'] = df['Content Rating'].str.lower().str.strip()
```

```python
df['Content Rating'] = df['Content Rating'].fillna('everyone')
```

```python
# Convert 'Installs' and 'Minimum Installs' to integers
df["Installs"] = df["Installs"].fillna(0).astype(int)
df["Minimum Installs"] = df["Minimum Installs"].fillna(0).astype(int)
```

```python
df.head(20)
```

```python
# حذف ردیف‌هایی که Rating آن‌ها خارج از بازه‌ی ۰ تا ۵ است
df = df[(df['Rating'] >= 0) & (df['Rating'] <= 5)]
```

```python
# حذف ردیف‌هایی که Installs آن‌ها منفی است
df = df[df['Installs'] >= 0]
```

```python
df = df.dropna(subset=["App Name"])
```

```python
df.info()
```

```python
df.to_csv('Google-Playstore.csv/cleaned_google_playstore_data.csv', index=False)
```

```python
# Load the preprocessed dataset
df = pd.read_csv("Google-Playstore.csv/cleaned_google_playstore_data.csv")
```

```python
# Save unique categories
categories = df[['Category']].drop_duplicates().reset_index(drop=True)
categories['id'] = categories.index + 1  # Assign unique IDs
categories.to_csv("categories.csv", index=False)
```

```python
# Save unique developers
developers = df[['Developer Id', 'Developer Website', 'Developer Email']].drop_duplicates().reset_index(drop=True)
developers['id'] = developers.index + 1  # Assign unique IDs
developers.rename(columns={'Developer Id': 'developer_id', 'Developer Website': 'website', 'Developer Email': 'email'}, inplace=True)
developers.to_csv("developers.csv", index=False)
```

```python
# Merge category and developer IDs back into the main dataset
df = df.merge(categories, on="Category", how="left")
df = df.merge(developers, left_on="Developer Id", right_on="developer_id", how="left")
```

```python
# Select relevant columns for the apps table
apps = df[['App Name', 'App Id', 'id_x', 'Rating', 'Rating Count', 'Installs',
           'Minimum Installs', 'Maximum Installs', 'Free', 'Price', 'Currency',
           'Size', 'Minimum Android', 'id_y', 'Released', 'Last Updated',
           'Content Rating', 'Privacy Policy', 'Ad Supported', 'In App Purchases', 'Editors Choice']]

# Rename columns to match database schema
apps.rename(columns={'id_x': 'category_id', 'id_y': 'developer_id', 'Size': 'size_mb'}, inplace=True)
# Save apps data to CSV
apps.to_csv("apps.csv", index=False)
```

```python
apps.head(2)
```

```python
# Convert 'Installs' and 'Minimum Installs' to integers
apps["Installs"] = apps["Installs"].fillna(0).astype(int)
apps["Minimum Installs"] = apps["Minimum Installs"].fillna(0).astype(int)

# Save to CSV
apps.to_csv("apps.csv", index=False)
```

```python
# Load the dataset
apps = pd.read_csv("apps.csv")

# Create a unique ID for each Content Rating
unique_ratings = apps["Content Rating"].dropna().unique()
rating_mapping = {rating: idx + 1 for idx, rating in enumerate(unique_ratings)}
rating_mapping
```

```python
#: Replace Content Rating strings with their corresponding IDs
apps["Content Rating"] = apps["Content Rating"].map(rating_mapping)

# Save the updated dataset with numeric Content Rating
apps.to_csv("apps_with_content_rating_id.csv", index=False)
apps
```

```python
# Save the mapping separately for reference
rating_df = pd.DataFrame(list(rating_mapping.items()), columns=["Content Rating", "id"])
rating_df.to_csv("content_rating.csv", index=False)
rating_df
```
